# 3.1. American Robin's Physical Dimensions (Bar Chart)

```python
# Data for the American Robin's physical attributes
robin_physical_data = pd.DataFrame({
    'Attribute': ['Length (cm)', 'Body Mass (g)', 'Wingspan (cm)'],
    'Min Value': [american_robin_data['MinLength'], american_robin_data['MinBodyMass'], american_robin_data['MinWingspan']],
    'Max Value': [american_robin_data['MaxLength'], american_robin_data['MaxBodyMass'], american_robin_data['MaxWingspan']]
})

# Reshape for plotting
robin_physical_data_melted = robin_physical_data.melt(id_vars='Attribute', var_name='Measurement Type', value_name='Value')

plt.figure(figsize=(10, 6))
sns.barplot(x='Attribute', y='Value', hue='Measurement Type', data=robin_physical_data_melted, palette='viridis')
plt.title(f'{chosen_bird_name} - Physical Characteristics (Min & Max)')
plt.ylabel('Value')
plt.xlabel('Attribute')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
