## Video Game Sales Analysis Notebook

### Introduction
This notebook explores the "Video Game Sales" dataset, sourced from Kaggle. The dataset contains a list of video games with sales figures, ratings, and other relevant information. My goal is to use this data to tell a story about the video game industry by answering questions like:
- What are the most common sales figures?
- How are video games distributed across different platforms?
- What's the distribution of game release years?
- How do different publishers perform in terms of sales?

I will use histograms to visualize the distribution of key variables and gain insights into the data.

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set a style for the plots for better visualization
sns.set_style("whitegrid")

# Load the dataset. Make sure the CSV file is in the same directory or accessible via Google Drive.
# For this example, let's assume the file is uploaded directly to Colab's temporary session storage.
try:
    df = pd.read_csv('vgsales.csv')
except FileNotFoundError:
    print("Error: 'vgsales.csv' not found. Please upload the dataset to the Colab environment.")
    # A placeholder dataframe for demonstration if the file is not found
    data = {'Rank': [1, 2, 3], 'Name': ['Wii Sports', 'Super Mario Bros.', 'Mario Kart Wii'], 'Platform': ['Wii', 'NES', 'Wii'], 'Year': [2006, 1985, 2008], 'Genre': ['Sports', 'Platform', 'Racing'], 'Publisher': ['Nintendo', 'Nintendo', 'Nintendo'], 'NA_Sales': [41.49, 29.08, 15.85], 'EU_Sales': [29.02, 3.58, 12.88], 'JP_Sales': [3.77, 6.81, 3.79], 'Other_Sales': [8.46, 0.77, 3.31], 'Global_Sales': [82.74, 40.24, 35.82]}
    df = pd.DataFrame(data)

# Display the first few rows of the dataframe to get a sense of the data
print("First 5 rows of the dataset:")
print(df.head())

# Get a summary of the data, including data types and non-null values
print("\nData information:")
print(df.info())

# Drop rows with missing values, specifically in the 'Year' column, as it's crucial for our analysis
df.dropna(subset=['Year'], inplace=True)
# Convert 'Year' column to integer type for proper histogram binning
df['Year'] = df['Year'].astype(int)

# Check the data types again to confirm the changes
print("\nData information after cleaning:")
print(df.info())
```

### Exploratory Data Analysis with Histograms

```python
#### Histogram 1: Distribution of Global Sales
# Let's see how the total global sales are distributed. We'll use a logarithmic scale to handle the wide range of values.
plt.figure(figsize=(10, 6))
plt.hist(df['Global_Sales'], bins=50, color='skyblue', edgecolor='black', log=True)
plt.title('Distribution of Global Sales (Log Scale)')
plt.xlabel('Global Sales (in millions)')
plt.ylabel('Number of Games (Log Scale)')
plt.show()
```

```python
#### Histogram 2: Distribution of Game Release Years
# This histogram shows us the number of games released over time.
plt.figure(figsize=(12, 7))
plt.hist(df['Year'], bins=30, color='lightgreen', edgecolor='black')
plt.title('Distribution of Video Game Release Years')
plt.xlabel('Year')
plt.ylabel('Number of Games Released')
plt.show()
```

```python
#### Histogram 3: Distribution of North American Sales
# Let's focus on the North American market to see its sales distribution.
plt.figure(figsize=(10, 6))
plt.hist(df['NA_Sales'], bins=50, color='salmon', edgecolor='black', log=True)
plt.title('Distribution of North American Sales (Log Scale)')
plt.xlabel('NA Sales (in millions)')
plt.ylabel('Number of Games (Log Scale)')
plt.show()
```

```python
#### Histogram 4: Distribution of European Sales
# Now, let's look at the European market's sales distribution for comparison.
plt.figure(figsize=(10, 6))
plt.hist(df['EU_Sales'], bins=50, color='lightcoral', edgecolor='black', log=True)
plt.title('Distribution of European Sales (Log Scale)')
plt.xlabel('EU Sales (in millions)')
plt.ylabel('Number of Games (Log Scale)')
plt.show()
```

```python
#### Histogram 5: Distribution of Games by Genre
# This is a bar chart of counts, which is often used for categorical data. Let's see the most popular genres.
plt.figure(figsize=(14, 7))
sns.countplot(y='Genre', data=df, order=df['Genre'].value_counts().index, palette='viridis')
plt.title('Number of Games per Genre')
plt.xlabel('Number of Games')
plt.ylabel('Genre')
plt.show()
```

### Conclusion
From our analysis using these histograms, we can draw a few key conclusions:
- The **global and regional sales distributions are heavily skewed**, with a vast majority of games selling relatively few copies. Only a small number of games achieve blockbuster status.
- The **number of games released per year peaked in the mid-2000s** and has since been in a slight decline, possibly due to a shift towards digital distribution and fewer large-scale physical releases.
- We can see a clear visual representation of the most common genres in the dataset, with **Action, Sports, and Shooter games** appearing most frequently.

This notebook successfully uses histograms and other plots to explore the dataset and tell a story about the distribution of video game sales, release years, and genres. It provides a solid foundation for further, more detailed analysis.
